```python
# Fix that remaining shit
for x in zillow:
    value = zillow[x].isna().sum()
    ratio = zillow[x].isna().sum() / len(zillow[x])
    print(f'\033[32m{x}\033[0m\n{value}\n{ratio}')
```

```python
# bathroomcnt.mode() = 2.0 (HAS FLOATS)
zillow.bathroomcnt = zillow.bathroomcnt.fillna(2.0)

# bedroomcnt.mode() = 3.0 (NO FLOATS)
zillow.bedroomcnt = zillow.bedroomcnt.fillna(3.0).astype(int)

# calculatedfinishedsquarefeet.mean() = 1831.44 (ONLY FLOAT)
zillow.calculatedfinishedsquarefeet = zillow.calculatedfinishedsquarefeet.fillna(1831.44)

# fips.mode() = 6037 (NO FLOATS)
zillow.fips = zillow.fips.fillna(6037).astype(int)

# fullbathcount.mode() = 2.0 (NO FLOATS)
zillow.fullbathcnt = zillow.fullbathcnt.fillna(2.0).astype(int)

# lotsizesquarefeet.mean() = 22604.34 (HAS FLOATS)
zillow.lotsizesquarefeet = zillow.lotsizesquarefeet.fillna(22604.34)

# regionidcity.mode() = 12447 (NO FLOATS)
zillow.regionidcity = zillow.regionidcity.fillna(12447).astype(int)

# roomcnt.mode() = 0 (NO FLOATS)
zillow.roomcnt = zillow.roomcnt.fillna(0).astype(int)

# yearbuilt.mode() = 1955 (NO FLOATS)
zillow.yearbuilt = zillow.yearbuilt.fillna(1955).astype(int)

# taxvaluedollarcnt.mean() = 443531.42 (ONLY FLOAT)
zillow.taxvaluedollarcnt = zillow.taxvaluedollarcnt.fillna(443531.42)

# assessmentyear.mode() = 2016 (NO FLOATS)
zillow.assessmentyear = zillow.assessmentyear.fillna(2016).astype(int)

# propertylandusedesc.mode() = 'Single Family Residential'
zillow.propertylandusedesc = zillow.propertylandusedesc.fillna('Single Family Residential')
```
